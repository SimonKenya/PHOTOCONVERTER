```python
#!/usr/bin/env python3

import os
from PIL import Image

# Define the source and destination directories
source_folder = '/home/students/images'
destination_folder = '/opt/icons/'

# Create the destination folder if it doesn't exist
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Iterate through each file in the source folder
for filename in os.listdir(source_folder):
    # Construct the full file path
    file_path = os.path.join(source_folder, filename)

    # Check if the file is an image
    if filename.lower().endswith(('.png', '.jpg', '.tiff', '.gif', '.bmp')):
        print(f"Processing file: {file_path}")
        # Open the image file
        with Image.open(file_path) as img:
            # Rotate the image 90° clockwise
            img = img.rotate(-90, expand=True)

            # Resize the image from 192x192 to 128x128
            img = img.resize((128, 128))

            # Define the new file path
            new_file_path = os.path.join(destination_folder, os.path.splitext(filename)[0] + '.jpeg')

            # Save the image in .jpeg format
            img.save(new_file_path, 'JPEG')
            print(f"Saved image to: {new_file_path}")
    else:
        print(f"Skipped file (not an image): {file_path}")
```

```python
#!/usr/bin/env python3

import os
import imghdr
from PIL import Image

# Define the source and destination directories
source_folder = '/home/student/images'
destination_folder = '/opt/icons/'

# Create the destination folder if it doesn't exist
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Iterate through each file in the source folder
for filename in os.listdir(source_folder):
    # Construct the full file path
    file_path = os.path.join(source_folder, filename)

    # Check if the file is an image and if it's a .tiff image
    if imghdr.what(file_path) == 'tiff':
        print(f"Processing file: {file_path}")
        # Open the image file
        with Image.open(file_path) as img:
            # Rotate the image 90° clockwise (to correct the anti-clockwise rotation)
            img = img.rotate(90, expand=True)

            # Resize the image to 128x128 pixels
            img = img.resize((128, 128))

            # Convert the image mode to 'RGB' if necessary
            if img.mode != 'RGB':
                img = img.convert('RGB')

            # Define the new file path
            new_file_path = os.path.join(destination_folder, os.path.splitext(filename)[0] + '.jpeg')

            # Save the image in .jpeg format
            img.save(new_file_path, 'JPEG')
            print(f"Saved image to: {new_file_path}")
    else:
        print(f"Skipped file (not a .tiff image): {file_path}")
```
